###1.In this task we have to find the students scores based on their study hours. This is a simple Regression problem type because it has only two variables.



```python
import pandas as pd

df = pd.DataFrame({'hours': [7.7,5.9,4.5,3.3,1.1,8.9,2.5,1.9,2.7,8.3,5.5,9.2,1.5,3.5,8.5,3.2,6.5],
                   'score': [79,60,45,33,12,87,21,19,29,81,58,88,14,34,85,32,66]})
df[0:18]
      
```

```python
import matplotlib.pyplot as plt

plt.scatter(df.hours, df.score)
plt.title('Hours studied vs. Exam Score')
plt.xlabel('Hours')
plt.ylabel('Score')
plt.show()
```

```python
df.boxplot(column=['score'])
```

```python
import statsmodels.api as sm
y = df['score']
x = df[['hours']]
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
print(model.summary())
```

```python
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_regress_exog(model, 'hours', fig=fig)
```

```python
res = model.resid
fig = sm.qqplot(res, fit=True, line="45")
plt.show() 
```
